```python
import cv2 as cv
import matplotlib.pyplot as plt
# 이미지 회전
img = plt.imread("C:\\1.jpg")
plt.imshow(img)
```

# 이미지 회전 (90, 180, 270)
### 130, 160 등 회전

그 외의 회전은 affine 값을 구하는등 계산식이 복잡함


또한 140도 회전등은 회전시 이미지의 소실 부분이 생겨
학습시 학습해야할 부분이 사라질 염려도 있음

만약 학습 할 이미지가 중앙에 있다면 130도 회전을 가해도 됨.

```python
img = plt.imread("C:\\1.jpg") 
img = cv.rotate(img, cv.ROTATE_180)
#img = cv.rotate(img, 1)
plt.imshow(img)
```

```python
img = plt.imread("C:\\1.jpg") 
img = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
#img = cv.rotate(img, 2)
plt.imshow(img)
```

```python
img = plt.imread("C:\\1.jpg") 
img = cv.rotate(img, cv.ROTATE_90_COUNTERCLOCKWISE)
#img = cv.rotate(img, 0)
plt.imshow(img)
```

## 이미지 flip

```python
img = plt.imread("C:\\1.jpg") 
img = cv.flip(img, 1)
plt.imshow(img)
```

```python
img = plt.imread("C:\\1.jpg") 
img = cv.flip(img, 0)
plt.imshow(img)
```

## 그림의 밝기 조절 

```python
# 밝기를 추가 (min = 0, max = 255)
import numpy as np
img = cv.imread("C:\\1.jpg",0) # 사진을 흑백으로 저장
rows, cols = img.shape[0],img.shape[1]
img_0 = np.zeros(shape=(rows,cols),dtype=np.uint8)
img_1 = np.full(shape=(rows,cols),fill_value=80, dtype=np.uint8)
img_2 = np.full(shape=(rows,cols),fill_value=160, dtype=np.uint8)

img0 = cv.add(img,img_0)
img1 = cv.add(img,img_1)
img2 = cv.add(img,img_2)

cv.imshow("add_0",img0)
cv.imshow("add_80",img1)
cv.imshow("add_160",img2)
cv.waitKey()
cv.destroyAllWindows()
```
